# Cek dan Cleaning Data

```python
# Ubah kolom tanggal menjadi format datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df
```

# Kolom Tambahan

```python
# Hitung lama pengiriman (dalam hari)
df['Durasi Pengiriman (hari)'] =  (df['Ship Date'] - df['Order Date']).dt.days

# Ekstrak bulan & tahun dari Order Date
df['Bulan'] = df['Order Date'].dt.month_name()
df['Tahun'] = df['Order Date'].dt.year
df['Bulan-Tahun'] = df['Order Date'].dt.to_period('M')
df
```

# Analisis Data 

```python
# Produk terlaris (berdasarkan kuantitas)
Produk_terlaris = df.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False).head(10)
print(Produk_terlaris)
```

```python
# Kategori paling menguntungkan (berdasarkan total penjualan)
kategori_terbaik = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(kategori_terbaik)
```

```python
# Wilayah dengan penjualan tertinggi
wilayah_terbaik = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(wilayah_terbaik)
```

```python
# Kota dengan penjualan tertinggi
kota_terbaik = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
print(kota_terbaik)
```

```python
# Durasi Pengiriman Rata-rata
durasi_rata = df['Durasi Pengiriman (hari)'].mean()
print(f"Rata-rata waktu pengiriman: {durasi_rata:.2f} hari")
```

```python
# Penjualan Bulanan
penjualan_bulanan = df.groupby('Bulan-Tahun')['Sales'].sum()
print(penjualan_bulanan)
```

```python
# Korelasi antar kolom numerik
corr = df[['Quantity', 'Sales', 'Durasi Pengiriman (hari)']].corr()
corr
```
